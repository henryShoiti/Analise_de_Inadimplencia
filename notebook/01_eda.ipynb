# Dicionário dos dados

### `SeriousDlqin2yrs`
Person experienced 90 days past due delinquency or worse. | Y/N

O cliente teve inadimplência de 90 ou mais dias nos últimos 2 anos?

### `RevolvingUtilizationOfUnsecuredLine` 
Total balance on credit cards and personal lines of credit except real estate and no installment debt like car loans divided by the sum of credit limits. | percentage

Saldo total em cartões de crédito e linhas de crédito pessoais, excluindo imóveis e dívidas não parceladas, como financiamentos de automóveis, dividido pela soma dos limites de crédito.

Quanto da linha de crédito rotativo o cliente está usando? (0 = nada, 1 = 100%)

### `age`
Age of borrower in years. | integer

Idade do cliente.

### `NumberOfTime30-59DaysPastDueNotWorse`
Number of times borrower has been 30-59 days past due but no worse in the last 2 years. | integer

Quantas vezes o cliente atrasou entre 30 e 59 dias nos últimos 2 anos.

### `DebtRatio`
Monthly debt payments, alimony,living costs divided by monthly gross income. | percentage

Pagamentos mensais de dívidas (dívidas, pensão alimentícia e despesas de subsistência) divididos pela renda bruta mensal.

### `MonthlyIncome`
Monthly income. | real

Renda mensal em dólares.

### `NumberOfOpenCreditLinesAndLoans`
Number of Open loans (installment like car loan or mortgage) and Lines of credit (e.g. credit cards). | integer

Número de linhas de crédito e empréstimos abertos.

### `NumberOfTimes90DaysLate`
Number of times borrower has been 90 days or more past due. | integer

Quantas vezes o cliente ficou 90+ dias em atraso.

### `NumberRealEstateLoansOrLine`
Number of mortgage and real estate loans including home equity lines of credit. | integer

Número de empréstimos hipotecários e imobiliários, incluindo linhas de crédito com garantia imobiliária.

### `NumberOfTime60-89DaysPastDueNotWorse`
Number of times borrower has been 60-89 days past due but no worse in the last 2 years. | integer

Quantas vezes o cliente atrasou entre 60 e 89 dias.

### `NumberOfDependents`
Number of dependents in family excluding themselves (spouse, children etc.). | integer

Número de dependentes (cônjuge, filhos, etc.).
